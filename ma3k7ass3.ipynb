```python
import random

#hat game func
def hat_game(n):
    hat = list(range(1, n + 1))

    while len(hat) > 1:
        a, b = random.sample(hat, 2)

        difference = abs(a - b)

        hat.remove(a)
        hat.remove(b)

        # if difference != 0:
        hat.append(difference)
    return hat[0]

#run 10 times
for x in range(10):
    result = hat_game(2025)
    print(f'The final remaining number is: {result}')
```

```python
import random
import matplotlib.pyplot as plt
from collections import Counter

#hat game func
def hat_game(n):
    hat = list(range(1, n + 1))

    while len(hat) > 1:
        a, b = random.sample(hat, 2)
        difference = abs(a - b)
        hat.remove(a)
        hat.remove(b)
        hat.append(difference)
    return hat[0]

#run many times
results = [hat_game(2025) for _ in range(10000)]

#max value
max_value = max(results)
print(f"The maximum value obtained over 10,000 runs is: {max_value}")

#plot
counter = Counter(results)

plt.figure(figsize=(12, 6))
plt.bar(counter.keys(), counter.values(), width=1.0, edgecolor="black")
plt.xlabel("Final Remaining Number")
plt.ylabel("Frequency")
plt.title("Distribution of Final Remaining Number in the Hat Game (n=2025)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import plotly.express as px
import pandas as pd
from collections import Counter

#plot the graph
counter = Counter(results)

min_val = min(results)
max_val = max(results)
possible_values = range(min_val, max_val + 1)

data = []
for value in possible_values:
    freq = counter.get(value, 0)
    data.append((value, freq))
    
df = pd.DataFrame(data, columns=['Final Remaining Number', 'Frequency'])

df['Final Remaining Number'] = df['Final Remaining Number'].astype(str)

fig = px.bar(
    df,
    x='Final Remaining Number',
    y='Frequency',
    title="Distribution of Final Remaining Number in the Hat Game (n=2025)",
    labels={'Final Remaining Number': 'Final Remaining Number', 'Frequency': 'Frequency'},
    # template="plotly_dark"
)
# fig.update_traces(marker_color='rgb(0,0,0)')  


fig.show()
```
