```python
import numpy as np
import matplotlib.pyplot as plt

def generate_ulam_spiral(n):
    spiral = np.zeros((n, n), dtype=int)
    x, y = n // 2, n // 2
    spiral[y, x] = 1
    number = 2
    step = 1
    while number <= n * n:
        for _ in range(step):
            if number > n * n:
                break
            x += 1
            spiral[y, x] = number
            number += 1
        for _ in range(step):
            if number > n * n:
                break
            y -= 1
            spiral[y, x] = number
            number += 1
        step += 1
        for _ in range(step):
            if number > n * n:
                break
            x -= 1
            spiral[y, x] = number
            number += 1
        for _ in range(step):
            if number > n * n:
                break
            y += 1
            spiral[y, x] = number
            number += 1
        step += 1
    return spiral

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def generate_prime_numbers_up_to(max_val):
    primes = set()
    for i in range(2, max_val + 1):
        if is_prime(i):
            primes.add(i)
    return primes

n = 255
spiral = generate_ulam_spiral(n)
prime_nums = generate_prime_numbers_up_to(n * n)

plt.figure(figsize=(20, 20))
ax = plt.gca()
plt.xlim(-0.5, n - 0.5)
plt.ylim(n - 0.5, -0.5)
ax.set_xticks(np.arange(-0.5, n, 1), minor=True)
ax.set_yticks(np.arange(-0.5, n, 1), minor=True)
# ax.grid(which='minor', color='gray', linestyle='-', linewidth=0.5)
ax.set_xticklabels([])
ax.set_yticklabels([])

for row in range(n):
    for col in range(n):
        val = spiral[row, col]
        if val in prime_nums:
            rect = plt.Rectangle((col - 0.5, row - 0.5), 1, 1, facecolor='red', edgecolor='none', alpha=1, zorder=1)
            ax.add_patch(rect)

plt.title('Spiral with Prime Numbers Shaded')
plt.show()
```
